### __3 задача__
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y = df['sqft_living'], kind='reg');
sns.jointplot(x=df['price'], y = df['sqft_lot'], kind='reg');
```

Взимосвязь цены на недвижимости в зависимости от жилой и общей площади. Наблюдается взаимосвязь, причем зависимость стоимости недвижимости от площади жилых помещений сильнее.

```python
df[df['sqft_lot']> 500000].sort_values(by='sqft_lot')
```

А зависимость цены от общей площади не так сильна, при чем цена недвижимости с максимальной общей площадью всего 700 000. 

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');
```

```python
sns.jointplot(x=df['lat'], y=df['price'], kind='reg');
```

Из графиков видно, что наибольшее скопление домов между широтой 47.5 и 47.7, по цене мы видим, что в зависимости от широты есть небольшая зависимость. Где- то после широты 47.5 стоимость домов увеличивается.

```python

df_cut = df[df['lat'] >= 47.5]
sns.jointplot(x=df_cut['lat'], y=df_cut['price'], kind='reg');
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(15,10))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

По матрице корреляций видим, что показателями, которые в влияют на увеличение цены являются жилая площадь и рейтинга дома(grade). Также небольшое влияние оказывает количество ванных комнат.

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.5)

plt.xlabel('Рейтинг дома')
plt.ylabel('Цена')
plt.title('Распределение цены в зависимости от рейтинга дома');
```

```python
# построим график без выбросов
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['grade'], y=df['price'], showfliers = False)

plt.xlabel('Рейтинг дома')
plt.ylabel('Цена')
plt.title('Распределение цены в зависимости от рейтинга дома');
```

```python
plt.figure(figsize=(6, 4))

sns.lineplot(x = df['bathrooms'], y = df['price'])

plt.title('Распределение цены в зависимости от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена')

plt.show();
```
